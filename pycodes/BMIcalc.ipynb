# simple Body Mass Index calculator


```python
weight = float(input("Enter the weight :"))
feet = int(input("Enter the feet :"))
inches = int(input("Enter the inches :"))
height=(feet * 12 + inches ) *0.0254
BMI_cal = weight / (height ** 2)
if BMI_cal < 18.25:
    print("Under Weight")
elif BMI_cal >= 18.25 and BMI_cal < 25:
    print("Normal")
elif BMI_cal >=25 and BMI_cal <30:
    print("Over Weight")
else:
    print("Obesity")
print("BMI :",round(BMI_cal,2))
```
