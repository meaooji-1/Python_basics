# number guessing game

```python
import random
aim_number = random.randint(1, 100)
guess = 0

print("a number between 1 and 100.")

while guess != aim_number:
    guess = int(input("Enter guess: "))
    
    if guess < aim_number:
        print("Too low")
    elif guess > aim_number:
        print("Too high")
    else:
        print("You win:")
```
